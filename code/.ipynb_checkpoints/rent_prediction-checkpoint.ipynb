# 2. Feature Engineering

```python
import pandas as pd

# Load and clean training data
train_data = pd.read_csv('./datasets/train_set.csv')
print('Data size before cleaning:', train_data.shape)
train_data = train_data.dropna(subset=['description', 'sqfeet', 'price'])
print('Data size after cleaning:', train_data.shape)

train_data.head(3)
```

```python
train_data.columns
```

```python
import numpy as np

# Create new features
train_data['desc_length'] = train_data['description'].str.len()
train_data['price_per_sqft'] = train_data['price'] / train_data['sqfeet'].replace(0, np.nan)

# Convert categorical features to numerical
categorical_columns = ['region', 'type', 'laundry_options', 'parking_options', 'state']
categorical_columns = [col for col in categorical_columns if col in train_data.columns]
train_data = pd.get_dummies(train_data, columns=categorical_columns, drop_first=True)

# Drop unnecessary features
train_data = train_data.drop(['id', 'description'], axis=1)

# Separate features and target
X_train = train_data.drop('price', axis=1)
y_train = np.log1p(train_data['price'].clip(lower=1))  # Ensure price >= 1 for log1p
```

# 3. Model Training

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer


# Split data into training and validation sets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

# Handle missing values and scale features
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()

X_train_split = pd.DataFrame(imputer.fit_transform(X_train_split), columns=X_train.columns)
X_val_split = pd.DataFrame(imputer.transform(X_val_split), columns=X_train.columns)
X_train_split = pd.DataFrame(scaler.fit_transform(X_train_split), columns=X_train.columns)
X_val_split = pd.DataFrame(scaler.transform(X_val_split), columns=X_train.columns)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_true, y_pred):
    y_pred_exp = np.expm1(y_pred)
    y_true_exp = np.expm1(y_true)

    mae = mean_absolute_error(y_true_exp, y_pred_exp)
    mse = mean_squared_error(y_true_exp, y_pred_exp)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true_exp, y_pred_exp)

    print("\nModel Evaluation Metrics:")
    print(f"{'Metric':<10} | {'Value':>10}")
    print("-" * 25)
    print(f"{'MAE':<10} | {mae:>10.2f}")
    print(f"{'MSE':<10} | {mse:>10.2f}")
    print(f"{'RMSE':<10} | {rmse:>10.2f}")
    print(f"{'R²':<10} | {r2:>10.4f}")
```

## GB model

```python
from sklearn.ensemble import GradientBoostingRegressor

# Train and evaluate the model
print("Training model...")
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train_split, y_train_split)

print("Evaluating model...")
y_pred = gb_model.predict(X_val_split)

evaluate_model(y_val_split, y_pred)
```

## KNN model

```python
from sklearn.neighbors import KNeighborsRegressor

# Train and evaluate the model
print("Training model...")
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_split, y_train_split)

print("Evaluating model...")
y_pred = knn_model.predict(X_val_split)

evaluate_model(y_val_split, y_pred)
```

# RF model

```python
from sklearn.ensemble import RandomForestRegressor

# Train and evaluate the model
print("Training model...")
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_split, y_train_split)

print("Evaluating model...")
y_pred = rf_model.predict(X_val_split)
evaluate_model(y_val_split, y_pred)
```

# 4. Model Inference

```python
import pandas as pd

# Load and clean test data
test_data = pd.read_csv('./datasets/test_set.csv')
print('Data size before cleaning:', test_data.shape)
test_data = test_data.dropna(subset=['description', 'sqfeet', 'price'])
print('Data size after cleaning:', test_data.shape)

test_data.columns
```

```python
test_data.head(3)
```

```python
ground_truth_price = test_data['price']

# Create new features
test_data['desc_length'] = test_data['description'].str.len()
test_data['price_per_sqft'] = test_data['price'] / test_data['sqfeet'].replace(0, np.nan)
test_data = pd.get_dummies(test_data, columns=categorical_columns, drop_first=True)

# Align test data with training data
missing_cols = set(X_train.columns) - set(test_data.columns)
for col in missing_cols:
    test_data[col] = 0
test_data = test_data[X_train.columns]

# Handle missing values and scale test data
test_data = pd.DataFrame(imputer.transform(test_data), columns=X_train.columns)
test_data = pd.DataFrame(scaler.transform(test_data), columns=X_train.columns)
```

```python
# Predict on test data for KNN model
test_pred_log_knn = knn_model.predict(test_data)
test_pred_price_knn = np.expm1(test_pred_log_knn)

# Evaluate KNN model
print("\nKNN Test Set Evaluation:")
evaluate_model(np.log1p(ground_truth_price), test_pred_log_knn)

# Predict on test data for Gradient Boosting model
test_pred_log_gb = gb_model.predict(test_data)
test_pred_price_gb = np.expm1(test_pred_log_gb)

# Evaluate Gradient Boosting model
print("\nGradient Boosting Test Set Evaluation:")
evaluate_model(np.log1p(ground_truth_price), test_pred_log_gb)

# Predict on test data for Random Forest model
test_pred_log_rf = rf_model.predict(test_data)
test_pred_price_rf = np.expm1(test_pred_log_rf)

# Evaluate Random Forest model
print("\nRandom Forest Test Set Evaluation:")
evaluate_model(np.log1p(ground_truth_price), test_pred_log_rf)
```

```python
# Combine the predictions into one DataFrame
comparison = pd.DataFrame({
    'Ground Truth Price': ground_truth_price,
    'KNN Predicted Price': test_pred_price_knn,
    'GB Predicted Price': test_pred_price_gb,
    'RF Predicted Price': test_pred_price_rf
})

print("\nSample Predictions from All Models:")
comparison.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
plt.plot(
    range(50), 
    comparison['Ground Truth Price'][:50], 
    label='Ground Truth Price', 
    marker='o', 
    linestyle='-', 
    color='black', 
    markersize=6, 
    linewidth=2
)
plt.plot(
    range(50), 
    comparison['KNN Predicted Price'][:50], 
    label='KNN Predicted Price', 
    marker='s', 
    linestyle='--', 
    color='blue', 
    markersize=5, 
    linewidth=1.5
)
plt.plot(
    range(50), 
    comparison['GB Predicted Price'][:50], 
    label='GB Predicted Price', 
    marker='^', 
    linestyle='-.', 
    color='green', 
    markersize=5, 
    linewidth=1.5
)
plt.plot(
    range(50), 
    comparison['RF Predicted Price'][:50], 
    label='RF Predicted Price', 
    marker='d', 
    linestyle=':', 
    color='red', 
    markersize=5, 
    linewidth=1.5
)

# Adding grid, labels, and title with customized fonts
plt.title('Model Predictions vs Ground Truth', fontsize=18, fontweight='bold')
plt.xlabel('Sample Index', fontsize=18)
plt.ylabel('Price', fontsize=18)
plt.legend(fontsize=12, loc='upper left', frameon=True, shadow=True, fancybox=True, borderpad=1)
plt.grid(color='gray', linestyle='--', linewidth=0.8, alpha=0.7)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)

# Add a bounding box around the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_linewidth(1.5)
plt.gca().spines['bottom'].set_linewidth(1.5)

plt.tight_layout()
plt.savefig('./figures/model_predictions_comparison.png', dpi=300, bbox_inches='tight')

plt.show()
```
